## 1.2 Créer des listes de mots.
La fonction suivante permet de renvoyer les mots d'une chaine de caractères comme élément d'une liste. Elle supprime tous les caractéres qui peuvent géner l'analyse comme les virgules, points virgule...

```python
def mot_dans_ligne(ligne) :
    ligne = ligne.replace(',','')
    ligne = ligne.replace('.','')
    ligne = ligne.replace(';','')
    ligne = ligne.replace(':','')
    ligne = ligne.replace("'",' ')
    ligne = ligne.replace('\n','')
    t=ligne.split(' ') # Separation des lignes en liste de mots, deux mots sont séparés par un espace
    return t
mot_dans_ligne("Ceci est la première ligne : à traiter.")
```

On peut l'utiliser pour lire les lignes de tout un fichier et stocker chaque ligne dans une liste :

```python
def mot_dans_fichier(fichier) :
    texte_entier=[]
    # Ouverture du fichier en lecture
    f = open(fichier,'r') 
    #Lecture de la premiere ligne
    ligne = f.readline()
    # Boucle permettant de parcourir toutes les lignes du fichier
    # tant que cette ligne existe
    while ligne:
        t = mot_dans_ligne(ligne)
        texte_entier.append(t)
        ligne = f.readline()
    return texte_entier
mot_dans_fichier('test.txt')
```

## 1.4 Un dictionnaire amélioré.
**Modifier le code de la fonction `dico-mot(liste)` pour le dictionnaire retourné ne contiennent que les mots dont la longueur est strictement supérieure à 2**

```python
# coding: utf-8
def dico_mot1(liste):
    dico,m,n,p = {},[],[],[]
    for i in liste:
        for j in i:
            m.append(j)
    for i in m:
        if i not in n:
            n.append(i)
    for i in m:
        if len(i)>2:
            p.append(i)
    for i in p:
        dico[i] = p.count(i)
    return dico
dico_test = dico_mot1(mot_dans_fichier('test.txt'))
print(dico_test)
```

## 2.3 Trouver le nombre d'occurences d'un mot
Utiliser un dictionnaire pour trouver le nombre d'occurence d'un mot dans le roman.

**Ecrire une fonction `occurence_mot(mot,dictionnaire)` qui renvoie le nombre d'occurences de `mot`.**

- Données :
    * `mot` : chaine de caractère resprésentant le mot dont on cherche le nombre d'occurences.
    * `dictionnaire` : dictionnaire dont les clés sont des mots (str) et dont les valeurs sont le nombre d'occurences du mot (int).
- Retourne : 
    * Entier, représentant le nombre d'occurence de `mot` dans le dictionnaire.
    * Retourne 0 si le mot n'est pas un clé du dictionnaire.

```python
def occurence_mot(mot,dictionnaire) :
    return dictionnaire[mot] if mot in dictionnaire else 0
```

```python
assert occurence_mot("Nana",dico_nana) == 820
assert occurence_mot("ordinateur",dico_nana) == 0
assert occurence_mot("a",dico_nana) == 0
assert occurence_mot("voiture",dico_nana) == 49
```

## 2.4 les mots les plus utilisés.
**Ecrire une fonction `util_mot(n,dictionnaire)` qui renvoie une liste contenant les mots dont $occurence>n$.**

- Données : 
    * `n` : Entier, qui représente le minimum d'occurences
    * `dictionnaire` : dictionnaire dont les clés sont des mots (str) et dont les valeurs sont le nombre d'occurences du mot (int).
- Retourne : Liste, dont les élèments sont les mot (str) dont le nombre d'occurences est supérieur à n.

```python
def util_mot(n,dictionnaire) :
    resultat =[]
    # YOUR CODE HERE
    for i,j in dictionnaire.items():
        if j >n :
            resultat.append(i.encode("latin1").decode())
    return resultat          
```
